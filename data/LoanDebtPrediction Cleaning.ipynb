# Datasets Cleaning

## Groceries Dataset Cleaning

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Groceries.csv')
df.head()
```

### Grocery Commodities Revenue

```python
pulses_revenue = []
for i in range(0, 1000):
    pulses_revenue.append((df['pulses_1kg'].loc[i])*(df['no_of_units_sold_per_day_of_pulses'].loc[i])*365)
pulses_revenue
```

```python
flour_revenue = []
for i in range(0, 1000):
    flour_revenue.append((df['flour_10kg'].loc[i])*(df['no_of_units_sold_per_day_of_flour'].loc[i])*365)
flour_revenue
```

```python
rice_revenue = []
for i in range(0, 1000):
    rice_revenue.append((df['flour_10kg'].loc[i])*(df['no_of_units_sold_per_day_of_rice'].loc[i])*365)
rice_revenue
```

```python
oil_revenue = []
for i in range(0, 1000):
    oil_revenue.append((df['flour_10kg'].loc[i])*(df['no_of_units_sold_per_day_of_oil'].loc[i])*365)
oil_revenue
```

```python
soap_revenue = []
for i in range(0, 1000):
    soap_revenue.append((df['flour_10kg'].loc[i])*(df['no_of_units_sold_per_day_of_soap'].loc[i])*365)
soap_revenue
```

### Total Revenue of Grocery Commodities

```python
total_revenue = []
for i in range(0, 1000):
    total_revenue.append(pulses_revenue[i] + flour_revenue[i] + rice_revenue[i] + oil_revenue[i] + soap_revenue[i])
total_revenue
```

```python
df['Total_Revenue'] = total_revenue
df
```

### Drop irrelevant columns

```python
df.drop(columns = ['Unnamed: 0','pulses_1kg', 'flour_10kg', 'rice_5kg', 'oil_1ltr', 'soap', 'no_of_units_sold_per_day_of_pulses', 'no_of_units_sold_per_day_of_flour', 'no_of_units_sold_per_day_of_rice', 'no_of_units_sold_per_day_of_oil', 'no_of_units_sold_per_day_of_soap'], inplace = True)
df
```
